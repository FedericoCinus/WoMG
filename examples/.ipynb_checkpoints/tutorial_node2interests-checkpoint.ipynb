```python
def read_emb(path, positive=pos):
    '''
    Returns embeddings from a node2vec file
    '''
    
    with open(path, 'r') as f:
        emb = f.readlines()
    #print(emb[0])
    embeddings = {}
    for i in range(1, 78):
        node = int(emb[i].split()[0])
        embeddings[node] = [float(emb[i].split()[j]) for j in range(1, 129)]
        
    #print(embeddings)
    if positive:
        minim = 0.
        for node in embeddings.keys():
            if min(embeddings[node]) < minim:
                minim = min(embeddings[node])
        for node in embeddings.keys():
            for i in range(len(embeddings[node])):
                embeddings[node][i] += abs(minim)   
        #print(embeddings[0])
    
    return embeddings
```

```python
def gformat(nx_obj):
    '''Sets the NetworkX graph in a dict format: tuple-link and int-weight
    '''
    graph_adj = nx.adjacency_matrix(nx_obj).toarray()
    G = {}
    for i in range(nx_obj.number_of_nodes()):
        for j in range(nx_obj.number_of_nodes()):
            if bool(graph_adj[i][j]):
                G[(i, j)] = graph_adj[(i, j)]
    #print(G)
    return G
```

The following analysis looks at the interests vectors distances between nodes that are connected and nodes that are not connected each other 

```python
def find_neigh(node, graph, out):
    '''out=False :Returns nodes list who follows the given node
       out=True :Returns nodes list followed by the given node
    '''
    if not out:
        lista = [link[0] for link in graph.keys() if link[1]==node and link[0]!=-1]
        #print(node, lista)
        return lista
    if out:
        lista = [link[1] for link in graph.keys() if link[0]==node]
        #print(node, lista)
        return lista
```

```python
def find_no_neigh(node, graph):
    '''Returns list of unfollowed and not following nodes for a given node
    '''
    unfollow_list = [i for i in range(77)]
    for n in find_neigh(node, graph, out=True):
        try:
            unfollow_list.remove(n)
        except ValueError:
            pass
    for n in find_neigh(node, graph, out=False):
        try:
            unfollow_list.remove(n) 
        except ValueError:
            pass
    #print(node, unfollow_list)
    return unfollow_list
```

We used different metrics, but for computational costs we suggest cosine similarity distance

```python
# metric 
def calc_cos_sim(node1, node2, interests):
    '''Returns scalar cosine similarity between two nodes vecs
       NB 0.0 value is the max similarity
    '''
    #print(interests)
    X = np.array(interests[node1])
    Y = np.array(interests[node2])
    dist = spatial.distance.cosine(X, Y)
    #print(node1, node2, dist)
    return dist
```

```python
def do_nmf(emb, components=15):
    '''
    Returns embeddings in a reducted dimensions space
    '''
    M = [] 
    for n in range(numb_nodes):
        M.append(emb[n])
    nmf = NMF(n_components=components)
    right = nmf.fit(M).components_
    left = nmf.transform(M)
    emb = {}
    for n in range(numb_nodes):
        emb[n] = left[n]
    return emb
```

```python
def divergence_stat(node, method, graph, interests):
    '''Returns statistics dict for the divergence of node interests
    '''
    # from graph
    follow = list(set(find_neigh(node, graph, out=True)) | set(find_neigh(node, graph, out=False)))
    #print(follow)
    unfollow = find_no_neigh(node, graph)
    '''
    if len(unfollow) > len(follow):
        rand_unfollow = random.sample(unfollow, len(follow))
    else:
        print('HI')
    '''
    rand_unfollow = unfollow
    
    # divergence
    follow_div = []
    for n in follow:
        if method == 'sim':
            follow_div.append(calc_cos_sim(node, n, interests))
    unfollow_div = []
    for n in rand_unfollow:
        if method == 'sim':
            unfollow_div.append(calc_cos_sim(node, n, interests))
    
    # stat
    stat = {}
    try:
        stat['unfollow mean'] = np.mean(unfollow_div)
        stat['follow mean'] = np.mean(follow_div)
        stat['unfollow max'] = np.max(unfollow_div)
        stat['follow max'] = np.max(follow_div)
        stat['unfollow min'] = np.min(unfollow_div)
        stat['follow min'] = np.min(follow_div)
        stat['unfollow std'] = np.std(unfollow_div)
        stat['follow std'] = np.std(follow_div)
    except:
        pass
    return stat
```

```python
def stats(data, method, reduction=False):
    '''Returns all statistics for the graph divergence study
    '''
    graph, interests = data
    if reduction:
        interests = do_nmf(interests)
    dic_stat = {}
    for n in range(numb_nodes):
        dic_stat[n] = divergence_stat(n, method, graph, interests)
    #datas
    #with open('dic_stat_'+str(method), 'wb') as f:
        #pickle.dump(dic_stat, f)
    
    stats = {}
    # means
    try: 
        stats['unfollow mean ' + method] = np.mean([dic_stat[n]['unfollow mean'] for n in range(numb_nodes)])
    except:
        pass
    try:
        stats['follow mean ' + method] = np.mean([dic_stat[n]['follow mean'] for n in range(numb_nodes)])
    except:
        pass
    try: 
        stats['unfollow SD ' + method] = np.std([dic_stat[n]['unfollow mean'] for n in range(numb_nodes)])
    except:
        pass
    try:
        stats['follow SD ' + method] = np.std([dic_stat[n]['follow mean'] for n in range(numb_nodes)])
    except:
        pass
    try:
        stats['unfollow max ' + method] = np.mean([dic_stat[n]['unfollow max'] for n in range(numb_nodes)])
    except:
        pass
    try:
        stats['follow max ' + method] = np.mean([dic_stat[n]['follow max'] for n in range(numb_nodes)])
    except:
        pass
    try:
        stats['unfollow min ' + method] = np.mean([dic_stat[n]['unfollow min'] for n in range(numb_nodes)])
    except:
        pass
    try:
        stats['follow min ' + method] = np.mean([dic_stat[n]['follow min'] for n in range(numb_nodes)])
    except:
        pass
    return stats
```

```python
def anal(reduction, idx=0):
    '''Analysis and plot function:
    FINDS EMBEDDINGS
    STATS OVER EMBEDDINGS
    '''
    
    #######################
    p = [0.25, 0.5, 1, 2, 4]
    q = [0.25, 0.5, 1, 2, 4]
    if idx==0:
        file_idx = [1,2,3,4,5]
        p = p[0]
    if idx==1:
        file_idx = [6,7,8,9,10]
        p = p[1]
    if idx==2:
        file_idx = [11,12,13,14,15]
        p = p[2]
    if idx==3:
        file_idx = [16,17,18,19,20]
        p = p[3]
    if idx==4:
        file_idx = [21,22,23,24,25]
        p = p[4]    
    #######################
    #analysis

    analysis = {}
    for i in file_idx:
        data = db[i]
        #print(data[1])
        #with open('dic_stat_'+str(i), 'w') as f:
        #    f.write(str(data[1]))
        analysis[i] = [stats(data, 'diff', reduction), stats(data, 'sim', reduction)]
        print('Finished parameters '+ str(i))

    ratio = list(reversed([p/j for j in q]))
    SD_follow_no_rescale_diff = [analysis[n][0]['follow SD diff'] for n in file_idx]
    SD_unfollow_no_rescale_diff = [analysis[n][0]['unfollow SD diff'] for n in file_idx]
    follow_no_rescale_diff = [analysis[n][0]['follow mean diff'] for n in file_idx]
    unfollow_no_rescale_diff = [analysis[n][0]['unfollow mean diff'] for n in file_idx]
    
    SD_follow_no_rescale_sim = [analysis[n][1]['follow SD sim'] for n in file_idx]
    SD_unfollow_no_rescale_sim = [analysis[n][1]['unfollow SD sim'] for n in file_idx]
    follow_no_rescale_sim = [analysis[n][1]['follow mean sim'] for n in file_idx]
    unfollow_no_rescale_sim = [analysis[n][1]['unfollow mean sim'] for n in file_idx]
    #plot
    
    
    cosine_dict = {'ratio': ratio, 'follow_no_rescale_sim': follow_no_rescale_sim, 'SD_follow_no_rescale_sim':SD_follow_no_rescale_sim,
                   'unfollow_no_rescale_sim': unfollow_no_rescale_sim, 'SD_unfollow_no_rescale_sim': SD_unfollow_no_rescale_sim
                  }
    
    return cosine_dict
```
